## **BROADCASTING**
- BROADCASTING IS A MECHANISM THAT ALLOWS NUMPY TO PERFORM OPERATION ON ARRAY  OF DIFFERNET SHAPE.
- THERE ARE MANY BROADCASTING RULEES WHICH YOU CAN LEARN FROM THE NUMPY DOCUMENTATION

```python
distance_in_km = np.array([5,2,7,3]) # shape : (4,)  dimension = 1
single_value = np.array(1000) # shape : ()  , dimension: 0
distance_in_km * single_value
```

```python
print(distance_in_km.shape)
print(single_value.shape)
```

```python
print(distance_in_km.ndim)
print(single_value.ndim)
```
